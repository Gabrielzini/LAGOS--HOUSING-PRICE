
### Baseline Model for  Data

The baseline model is created using the mean value of the test data. This serves as a benchmark to compare the performance of our regression models.

#### Steps:
1. Calculate the mean of the target variable (`y_test`) in the test dataset.
2. Use this mean value as the predicted value for all test samples.
3. Evaluate the baseline model using the Root Mean Squared Error (RMSE) metric.

#### Purpose:
The baseline model provides a simple reference point. Any regression model should ideally perform better than this baseline to be considered effective.


````python
# Baseline model using mean of training set
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# Calculate baseline predictions
y_pred_base_train = np.repeat(y_train.mean(), len(y_train))
y_pred_base_test = np.repeat(y_train.mean(), len(y_test))

# Calculate baseline metrics
base_r2_train = r2_score(y_train, y_pred_base_train)
base_mae_train = mean_absolute_error(y_train, y_pred_base_train)
base_mae_test = mean_absolute_error(y_test, y_pred_base_test)
base_rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_base_test))

print(f"Baseline R-square (training): {base_r2_train:.4f}")
print(f"Baseline Model MAE (training): ₦{base_mae_train:,.2f}")
print(f"Baseline Model MAE (test): ₦{base_mae_test:,.2f}")
print(f"Baseline Model RMSE (test): ₦{base_rmse_test:,.2f}")

# Store baseline predictions for later use
base_pred = y_pred_base_test
````

## 1.1 LINEAR REGRESSION 

````python
# Linear regression model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Predicting on the test set
y_pred = lin_reg.predict(X_test)

# Calculate metrics
lin_MAE = mean_absolute_error(y_test, y_pred)   
lin_RMSE = np.sqrt(mean_squared_error(y_test, y_pred))
lin_r2 = r2_score(y_test, y_pred)

print('Linear Regression Results:')
print(f'RMSE: ₦{lin_RMSE:,.2f}')
print(f'MAE: ₦{lin_MAE:,.2f}')
print(f'R²: {lin_r2:.4f}')

# Cross-validation
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_validate

# Perform cross-validation
cv_results = cross_validate(lin_reg, X, y, cv=5, 
                           scoring=['neg_root_mean_squared_error', 'neg_mean_absolute_error', 'r2'])

lin_cv_rmse = -cv_results['test_neg_root_mean_squared_error']
lin_cv_mae = -cv_results['test_neg_mean_absolute_error']
lin_cv_r2 = cv_results['test_r2']

print('\nCross-Validation Results:')
print(f'RMSE: ₦{lin_cv_rmse.mean():,.2f} ± {lin_cv_rmse.std():,.2f}')
print(f'MAE: ₦{lin_cv_mae.mean():,.2f} ± {lin_cv_mae.std():,.2f}')
print(f'R²: {lin_cv_r2.mean():.4f} ± {lin_cv_r2.std():.4f}')



````

````python
# Plotting the predictions vs actual values with additional diagnostic plots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# 1. Actual vs Predicted Price
axes[0, 0].scatter(y_test/1e6, y_pred/1e6, alpha=0.5)
axes[0, 0].plot([y_test.min()/1e6, y_test.max()/1e6], [y_test.min()/1e6, y_test.max()/1e6], 'k--', lw=2)
axes[0, 0].set_xlabel('Actual Price (Million ₦)')
axes[0, 0].set_ylabel('Predicted Price (Million ₦)')
axes[0, 0].set_title('Actual vs Predicted Price')
axes[0, 0].grid(True, alpha=0.3)

# 2. Residual Plot
residuals = y_test - y_pred
axes[0, 1].scatter(y_pred/1e6, residuals/1e6, alpha=0.5)
axes[0, 1].axhline(y=0, color='r', linestyle='--', lw=2)
axes[0, 1].set_xlabel('Predicted Price (Million ₦)')
axes[0, 1].set_ylabel('Residuals (Million ₦)')
axes[0, 1].set_title('Residual Plot')
axes[0, 1].grid(True, alpha=0.3)

# 3. Absolute Error Plot
abs_errors = np.abs(residuals)
axes[1, 0].scatter(y_pred/1e6, abs_errors/1e6, alpha=0.5, color='orange')
axes[1, 0].set_xlabel('Predicted Price (Million ₦)')
axes[1, 0].set_ylabel('Absolute Error (Million ₦)')
axes[1, 0].set_title('Absolute Error vs Predicted Price')
axes[1, 0].grid(True, alpha=0.3)

# 4. Histogram of Residuals
axes[1, 1].hist(residuals/1e6, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
axes[1, 1].set_xlabel('Residuals (Million ₦)')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].set_title('Distribution of Residuals')
axes[1, 1].grid(True, alpha=0.3)
axes[1, 1].axvline(x=0, color='r', linestyle='--', lw=2)

plt.tight_layout()
plt.show()

# Print some diagnostic statistics
print("\nDiagnostic Statistics:")
print(f"Mean Residual: ₦{residuals.mean():,.2f}")
print(f"Std Residual: ₦{residuals.std():,.2f}")
print(f"Mean Absolute Error: ₦{abs_errors.mean():,.2f}")
print(f"Max Absolute Error: ₦{abs_errors.max():,.2f}")
print(f"Min Absolute Error: ₦{abs_errors.min():,.2f}")
````
